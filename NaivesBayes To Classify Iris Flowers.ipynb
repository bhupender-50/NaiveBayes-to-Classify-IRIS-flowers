```python
features=data[['sepal_length','sepal_width','petal_length','petal_width']]
```

```python
target=data['species']
```

```python
target.value_counts()
```

# Converting Text value to Numerical type

```python
from sklearn.preprocessing import LabelEncoder
```

```python
target=pd.Series(LabelEncoder().fit_transform(target))
```

```python
target.value_counts()
```

# Visualizing the data

```python
import seaborn as sns
sns.pairplot(data, hue='species', vars=['sepal_length','sepal_width','petal_length','petal_width'])
```

```python
sns.countplot(data['species'], label = "Count")
```

```python
sns.heatmap(data.corr(), annot=True)
```

#### Splitting data into train and test data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(features,target,test_size=0.15, random_state=42)
```

# Importing Naives Bayes & Prediction

```python
from sklearn.naive_bayes import GaussianNB
```

```python
clf=GaussianNB()
clf.fit(X_train,Y_train)
```

```python
pred=clf.predict(X_test)
print(pred)
```

# Classification Report

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(pred,Y_test))
```
